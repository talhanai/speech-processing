```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
# filepath
# origin: https://catalog.ldc.upenn.edu/LDC93S1
file = 'data/LDC93S1.wav'

# desired sampling rate of audio
sample_rate = 16000

# load audio
audio, sample_rate = librosa.load(file, sample_rate)
```

```python
# load word transciption
words = []

with open('data/LDC93S1.wrd', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ')
    for row in reader:
        print(row[0], row[1], row[2])
        words.append(row)
```

```python
# load phone transciption
phones = []

with open('data/LDC93S1.phn', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ')
    for row in reader:
        print(row[0], row[1], row[2])
        phones.append(row)
```

```python
fig = plt.figure(figsize=(14,4))
plt.plot(audio)

plt.xlabel('samples')
plt.title('Audio "she had your dark suit in greasy wash water all year"')
plt.savefig('figures/plot-1.png')
plt.show()
```

```python
fig = plt.figure(figsize=(14,4))
plt.plot(audio)
for word in words:
    w_start = int(word[0])
    w_end   = int(word[1])
    w       = word[2]
    plt.text(x=w_start, y=np.min(audio),s=w, fontsize=12)
    plt.vlines(x=w_start, ymin=np.min(audio), ymax=np.max(audio), alpha=0.5)

plt.xlabel('samples')
plt.title('Audio "she had your dark suit in greasy wash water all year"')
plt.savefig('figures/plot-2.png')
plt.show()
```

```python
# DEFINING VARIABLES
# This will be used throughout the notebook

# [hop_length] define window shift of 10ms (typical)
shift = 0.01
hop_length = int(sample_rate * shift)

# [win_length] window size of 20ms (typical)
window = 0.02
win_length = int(sample_rate * window)

# [power] then squared
power = 2

# [n_mels] then binned over 128 mel sized triangular bins
n_mels = 128
```

```python
# PLOT WITH LIBROSA
# This serves as an example of plotting with librosa, 
# however you may find it easier to use matplotlib (an example in the next section).

# Calculating spectrogram
S = librosa.feature.melspectrogram(y=audio, sr=sample_rate, n_mels=n_mels, fmax=8000, power=power,
                                   hop_length=hop_length, win_length=win_length)


# Converting Power to DB
S_dB = librosa.power_to_db(S, ref=np.max)

plt.figure(figsize=(14,4))

# Plot spectrogram with librosa
librosa.display.specshow(S_dB, hop_length=hop_length, sr=sample_rate, x_axis='time', y_axis='mel')

plt.savefig('figures/plot-3.png')
plt.show()
```

```python
# PLOT SPECTROGRAM WITH MATPLOTLIB
# You may find this easier to work with than librosa display
# since you can manipulate the plot yourself.

# define figure size
plt.figure(figsize=(14,4))

# plot spectrogram
plt.imshow(S_dB, origin='lower', aspect='auto')

# define yaxis tick labels
ylocs = np.arange(0,n_mels,16).tolist()
freq  = np.round(librosa.mel_frequencies(n_mels=n_mels))[ylocs].tolist()
plt.yticks(ylocs,freq)

# define xaxis tick labels
timestep = 0.5 # seconds
xlocs = (np.arange(0,S_dB.shape[1], timestep / shift)).tolist()
xvals = (np.arange(0,audio.shape[0] / sample_rate, timestep)).tolist() 
plt.xticks(xlocs,xvals)

# colorbar
cbar = plt.colorbar()
cbar.ax.set_title('dB')

# plot labels
plt.xlabel('time (seconds)')
plt.ylabel('frequency (Hz)')
plt.title('Audio "she had your dark suit in greasy wash water all year"')

plt.tight_layout()
plt.savefig('figures/plot-4.png')
plt.show()
```

```python
# PLOTTING WORD LEVEL ALIGNMENTS WITH SPECTROGRAM
fig = plt.figure(figsize=(14,4))

# plot spectrogram
plt.imshow(S_dB, origin='lower', aspect='auto')

# define yaxis tick labels
ylocs = np.arange(0,n_mels,16).tolist()
freq  = np.round(librosa.mel_frequencies(n_mels=n_mels))[ylocs].tolist()
plt.yticks(ylocs,freq)

# define xaxis tick labels
xlocs = (np.arange(0,S_dB.shape[1], timestep / shift)).tolist()
xvals = (np.arange(0,audio.shape[0] / sample_rate,timestep)).tolist() 
plt.xticks(xlocs,xvals)

# colorbar
cbar = plt.colorbar()
cbar.ax.set_title('dB')

# plot labels
plt.xlabel('time (seconds)')
plt.ylabel('frequency (Hz)')
plt.title('Word-aligned transcription')

# plot word alignments
for word in words:
    
    # convert samples into spectrogram x-axis space
    w_start = (int(word[0]) / sample_rate) / shift
    w_end   = (int(word[1]) / sample_rate) / shift
    w       = word[2]
    
    # plot word text
    plt.text(x=w_start, y=n_mels-7,s=w, fontsize=12, color="white")
    
    # plot start/end of word
    plt.vlines(x=w_start, ymin=0, ymax=n_mels-1, alpha=0.5, color="white")


# save figure
plt.tight_layout()
plt.savefig('figures/plot-5.png')
plt.show()
```

```python
# PLOTTING PHONE LEVEL ALIGNMENTS WITH SPECTROGRAM
fig = plt.figure(figsize=(14,4))

# plot spectrogram
plt.imshow(S_dB, origin='lower', aspect='auto')

# define yaxis tick labels
ylocs = np.arange(0,n_mels,16).tolist()
freq  = np.round(librosa.mel_frequencies(n_mels=n_mels))[ylocs].tolist()
plt.yticks(ylocs,freq)

# define xaxis tick labels
xlocs = (np.arange(0,S_dB.shape[1], timestep / shift)).tolist()
xvals = (np.arange(0,audio.shape[0] / sample_rate, timestep)).tolist() 
plt.xticks(xlocs,xvals)

# colorbar
cbar = plt.colorbar()
cbar.ax.set_title('dB')

# plot labels
plt.xlabel('time (seconds)')
plt.ylabel('frequency (Hz)')
plt.title('Phone-Aligned transcription')

# plot word alignments
for i, phone in enumerate(phones):
    
    # convert the data into samples in the spectrogram
    ph_start = (int(phone[0]) / sample_rate) / 0.01
    ph_end   = (int(phone[1]) / sample_rate) / 0.01
    ph       = phone[2]
    
    # jiggle/space vertically the position of the phone text
    # so that they don't get printed on top of each other
    dip      = (i % 3) * 8
    
    # plot alignment text
    plt.text(x=ph_start, y=n_mels-5-dip,s=ph, fontsize=12, color="white")
    
    # plot start/end point of segment
    plt.vlines(x=ph_start, ymin=0, ymax=n_mels-1, alpha=0.5, color="white")


plt.tight_layout()
plt.savefig('figures/plot-6.png')
plt.show()
```
