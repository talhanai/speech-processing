```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
# filepath
# origin: https://catalog.ldc.upenn.edu/LDC93S1
file = 'data/LDC93S1.wav'

# desired sampling rate of audio
sample_rate = 16000

# load audio
audio, sample_rate = librosa.load(file, sample_rate)
```

```python
# plot audio signal
plt.plot(audio)
```

```python

# [hop_length] define window shift of 10ms
hop_length = int(sample_rate * 0.01)

# [win_length] window size of 20ms
win_length = int(sample_rate * 0.02)

# [power] then squared
power = 2

# [n_mels] then binned over 128 mel sized triangular bins
n_mels = 128

# Calculating spectrogram
S = librosa.feature.melspectrogram(y=audio, sr=sample_rate, n_mels=n_mels, fmax=8000, power=power,
                                   hop_length=hop_length, win_length=win_length)


# Converting Power to DB
S_dB = librosa.power_to_db(S, ref=np.max)

print(' Number of mel filters:', S_dB.shape[0])
print(' Number of windows:', S_dB.shape[1])
```

```python
# Plot spectrogram
librosa.display.specshow(S_dB, hop_length=hop_length, x_axis='time', y_axis='mel')
```
